```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

experiment1 = ['s', 's', 'b', 'b', 's', 's', 's', 's', 'b', 's']
nrCap1 = experiment1.count('s')
nrPajura1 = experiment1.count('b')

alpha1 = 1
beta1 = 1

alphaUpdate1 = alpha1 + nrCap1
betaUpdate1 = beta1 + nrPajura1

p = np.linspace(0, 1, 500)
dist1 = beta.pdf(p, alphaUpdate1, betaUpdate1)

plt.figure()
plt.plot(p, dist1, label=f'Posterior Beta({alphaUpdate1}, {betaUpdate1})')
plt.title('Posterior dupa primul experiment')
plt.xlabel('P(cap)')
plt.ylabel('Densitate')
plt.legend()
plt.grid(True)
plt.show()

experiment2 = ['b', 'b', 's', 'b', 's', 's', 'b', 's', 's', 'b']
nrCap2 = experiment2.count('s')
nrPajura2 = experiment2.count('b')

alpha2a = 1
beta2a = 1

alphaUpdate2a = alpha2a + nrCap2
betaUpdate2a = beta2a + nrPajura2

dist2a = beta.pdf(p, alphaUpdate2a, betaUpdate2a)

alpha2b = alphaUpdate1
beta2b = betaUpdate1

alphaUpdate2b = alpha2b + nrCap2
betaUpdate2b = beta2b + nrPajura2

dist2b = beta.pdf(p, alphaUpdate2b, betaUpdate2b)

plt.figure()
plt.plot(p, dist2a, label=f'Prior Beta Uniform: Beta({alphaUpdate2a}, {betaUpdate2a})')
plt.plot(p, dist2b, label=f'Prior din primul experiment: Beta({alphaUpdate2b}, {betaUpdate2b})')
plt.title('Posterior dupa al doilea experiment')
plt.xlabel('P(cap)')
plt.ylabel('Densitate')
plt.legend()
plt.grid(True)
plt.show()
```
